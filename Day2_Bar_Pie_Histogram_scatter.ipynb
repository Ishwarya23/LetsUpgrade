Bar plot

```python
x = [2,8,10]
y = [11,16,9]
x2 = [3,9,11]
y2 = [6,15,7]
plt.barh(x,y,label='x-y comparision',height=.3)
plt.barh(x2,y2,label='x2-y2 comparision')
plt.legend()
plt.show()
```

```python
x = [2,8,10]
y = [11,16,9]
x2 = [3,9,11]
y2 = [6,15,7]
plt.bar(x,y,label='x-y comparision',width=.3)
plt.bar(x2,y2,label='x2-y2 comparision')
plt.legend()
plt.show()
```

Pie chart

```python
labels = 'python','c++','Ruby','Java'
sizes = [215,130,245,210]
colors = ['gold' ,'yellowgreen','lightcoral','lightyellow']
plt.pie(sizes,labels=labels,colors=colors)
```

```python
labels = 'python','c++','Ruby','Java'
sizes = [215,130,245,210]
colors = ['gold' ,'yellowgreen','lightcoral','lightyellow']
plt.pie(sizes,labels=labels,colors=colors,autopct='%1.1f%%',shadow=False)
```

```python
labels = 'python','c++','Ruby','Java'
sizes = [215,130,245,210]
colors = ['gold' ,'yellowgreen','lightcoral','lightyellow']
explode = (0.4,0,0,0)
plt.pie(sizes,labels=labels,colors=colors,explode=explode,autopct='%1.1f%%',shadow=False)
```

```python
df = pd.Series([0.2,0.3,0.1,0.6],index=['a','b','c','d'],name='Pie plot')
df.plot.pie(autopct='%1.1f%%')
```
