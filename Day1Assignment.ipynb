Question 4

Plot a line plot showing the sales trend in company 1 and 2:

days = [1,2,3,4,5,6,7] #days of d week
sales_1 = [160,150,140,145,175,165,180] #sales of company1
sales_2 = [70,90,160,150,140,145,175]  #sales of company2

```python
days = [1,2,3,4,5,6,7]
sales_1 = [160,150,140,145,175,165,180] 
sales_2 = [70,90,160,150,140,145,175]
plt.plot(days,sales_1,'go-',label='Sales of company 1',markersize=5)
plt.plot(days,sales_2,'b*--',label='Sales of company 2',markersize=5)
plt.xlabel('Days')
plt.ylabel('Sales of company')
plt.title('Sales comparision')
plt.legend()
```

Question 5

Create a 3 by 3 subplots:
multiple plots
x = [1,2,3,4]
y1 = [4,3,2,1]
y2 = [10,20,30,40]
y3 = [40,30,20,10]

```python
x = [1,2,3,4]
y1 = [4,3,2,1]
y2 = [10,20,30,40]
y3 = [40,30,20,10]
y4 = [1,2,1,2]
y5 = [40,70,90,70]


plt.subplot(2,3,1)
plt.plot(x,y1)
plt.title('Plot1')
plt.subplot(2,3,2)
plt.plot(x,y2)
plt.title('Plot2')
plt.subplot(2,3,3)
plt.plot(x,y3)
plt.title('Plot3')
plt.subplot(2,3,4)
plt.plot(x,y4)
plt.title('Plot4')
plt.subplot(2,3,5)
plt.plot(x,y5)
plt.title('Plot5')
```
