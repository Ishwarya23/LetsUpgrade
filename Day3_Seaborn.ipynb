```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
```

```python
sns.get_dataset_names()
```

```python
tipdata = sns.load_dataset('tips')
tipdata.head()
```

relplot

```python
sns.relplot(x='total_bill',y='tip',data=tipdata,hue='smoker',style='time')
```

```python
sns.relplot(x='total_bill',y='tip',data=tipdata,hue='smoker',col='time')
```

```python
from numpy.random import randint,randn
df = pd.DataFrame(dict(Col1 = np.arange(500),Col2 = randn(500).cumsum()))
df           
```

```python
df.shape
```

```python
sns.relplot(x='Col1',y='Col2',kind='line',data=df)
```

```python
sns.relplot(x='Col1',y='Col2',kind='line',data=df,sort=True)
```

ScatterPlot

```python
sns.scatterplot(x='total_bill',y='tip',data=tipdata,size='size')
```

Categorical data plotting

catplot

```python
sns.catplot(x='day',y='total_bill',data=tipdata)
```

```python
sns.catplot(x='day',y='total_bill',data=tipdata,kind='box',hue='time')
```

```python
sns.catplot(x='day',y='total_bill',data=tipdata,kind='box',hue='time',dodge=False)
```

```python
sns.catplot(x='day',y='total_bill',data=tipdata,kind='box',hue='time')
```

```python
sns.catplot(x='day',y='total_bill',data=tipdata,kind='swarm',hue='time')
```

```python
sns.catplot(x='day',y='total_bill',data=tipdata,kind='boxen',hue='time')
```

```python
sns.catplot(x='day',y='total_bill',data=tipdata,kind='violin',hue='time')
```
