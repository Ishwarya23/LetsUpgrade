#Line Plot

```python
x=np.arange(1,11)
y=np.arange(11,21)
plt.title('LinePlot')
plt.plot(x,y,'go--',linewidth=2,markersize=5)
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

```python
x = np.arange(1,11)
y = 3 * x+5
plt.title('LinePlot')
plt.plot(x,y,'b*--',linewidth=2,markersize=6)
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

```python
x = np.arange(0, 4*np.pi,0.1)
y = np.sin(x)
plt.title('Sine waveform')
plt.plot(x,y,color = 'r')
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

#Subplot

```python
plt.subplot(2,2,1)
plt.plot(x,y,'g^--',markersize=2)
plt.subplot(2,2,2)
plt.plot(x,y,'r*-',markersize=2)
plt.subplot(2,2,3)
plt.plot(x,y,'bo--',markersize=2)
plt.subplot(2,2,4)
plt.plot(x,y,'r^--',markersize=2)
```
