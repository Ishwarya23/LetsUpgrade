```python
from matplotlib import pyplot as plt
import numpy as np
```

Plot 

```python
x = [1,2,3]
y = [5,6,7]
plt.plot(x,y,'r*-',label='Line',LineWidth=3)
plt.legend()
plt.title("Graph..")
plt.xlabel("X AXIS")
plt.ylabel("Y AXIS")
plt.grid()
plt.show()
```

Bar graph

```python
plt.bar([0.25,1.25,2.25,3.25,4.25],[50,40,70,80,20],label="BMW",color='r',width=.3)
plt.bar([.75,1.75,2.75,3.75,4.75],[80,20,20,50,60],label="Audi", color='b',width=.5)
plt.legend()
plt.show()
```

Histogram

```python
mark=[90,91,23,45,67,89,89,99,97,98,23,23,23,45,45]
#bins = np.arange(20,100)
bins = [20,30,40,50,60,70,80,90,100]
plt.hist(mark,bins,histtype='bar')
plt.show()
```

Scatter Plot

```python
x = [1,1.5,2,2.5,3,3.5,3.6]
y = [7.5,8,8.5,9,9.5,10,10.5]
x1=[8,8.5,9,9.5,10,10.5,11]
y1=[3,3.5,3.7,4,4.5,5,5.2]
plt.scatter(x,y, label='high income low saving',color='r')
plt.scatter(x1,y1,label='low income high savings',color='b')
plt.xlabel('saving*100')
plt.ylabel('income*1000')
plt.title('Scatter Plot')
plt.legend()
plt.show()
```

creating multiple plots in single line

```python
x = np.arange(0,20,0.5)
plt.plot(x,x**2,'g--',x,x,'ro-')
```

Area plot

```python
days = [1,2,3,4,5]
sleeping =[7,8,6,11,7]
eating = [2,3,4,3,2]
working =[7,8,7,2,2]
playing = [8,5,7,8,13]
plt.plot([],[],color='m', label='Sleeping', linewidth=5)
plt.plot([],[],color='c', label='Eating', linewidth=5)
plt.plot([],[],color='r', label='Working', linewidth=5)
plt.plot([],[],color='k', label='Playing', linewidth=5)
plt.stackplot(days, sleeping,eating,working,playing,colors=['m','c','r','b'])
plt.xlabel('x')
plt.ylabel('y')
plt.title('Stack Plot')
plt.legend()
plt.show()
```

Pie chart - circular graph

```python
slices = [2,2,2,13]
activities = ['sleeping','eating','working','playing']
cols = ['c','m','r','b']
plt.pie(slices,labels=activities,colors=cols,startangle=180  )
plt.title('Pie Plot')
plt.show()
```
